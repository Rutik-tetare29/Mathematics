### 19/1/25 -- LIMITS IN CALCULUS

```python
from sympy import Symbol, limit

# Define the variable and function
x = Symbol('x')
expr = (x**2 - 4) / (x - 2)

# Calculate the limit as x approaches 2
result = limit(expr, x, 2)
print("Limit of the function as x approaches 2:", result)
```

### 20/1/25 -- DERIVATIVES (BASICS)

```python
from sympy import diff

# Define the variable and function
x = Symbol('x')
function = x**3 + 2*x 

# Compute the derivative
derivative = diff(function, x)
print("Derivative of the function:", derivative)
```

### Applications of Derivatives

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
x = np.linspace(-10, 10, 400)
y = x**2
dy = 2 * x  # Derivative of x^2

# Plot the function and its derivative
plt.plot(x, y, label="f(x) = x^2", color="blue")
plt.plot(x, dy, label="f'(x) = 2x (Derivative)", color="red")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.title("Function and its Derivative")
plt.show()
```

### 21/1/25 -- PARTIAL DERIVATIVES

```python
from sympy import symbols, diff

# Define the variables and function
x, y = symbols('x y')
function = x**2 * y + y**3

# Compute partial derivatives
partial_x = diff(function, x)  # Partial derivative w.r.t x
partial_y = diff(function, y)  # Partial derivative w.r.t y

print("Partial derivative w.r.t x:", partial_x)
print("Partial derivative w.r.t y:", partial_y)
```

### 21/1/25 -- INTEGRATION (BASICS)

```python
from sympy import integrate

# Define the variable and function
x = Symbol('x')
function = x**2

# Compute indefinite integral
indefinite_integral = integrate(function, x)
print("Indefinite integral:", indefinite_integral)

# Compute definite integral from 0 to 2
definite_integral = integrate(function, (x, 0, 2))
print("Definite integral from 0 to 2:", definite_integral)
```

### 22/1/25 -- NUMERICAL INTEGRATION

```python
from scipy.integrate import quad

# Define the function
def func(x):
    return x**2

# Compute the integral from 0 to 2
result, error = quad(func, 0, 2)
print("Numerical integration result:", result)  # Output: 2.666...
```

### Applications of Integration

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def func(x):
    return x**2

# Compute the area
area, _ = quad(func, 1, 3)

# Plot the function and shade the area
x = np.linspace(0, 4, 400)
y = func(x)
plt.plot(x, y, label="f(x) = x^2")
plt.fill_between(x, y, where=(x >= 1) & (x <= 3), color="skyblue", alpha=0.5, label=f"Area = {area:.2f}")
plt.legend()
plt.show()
```

### 23/1/25 -- GRADIENT AND HESSIAN

```python
from sympy import hessian

# Define the variables and function
x, y = symbols('x y')
function = x**2 + y**2

# Compute the gradient
gradient = [diff(function, var) for var in (x, y)]
print("Gradient:", gradient)

# Compute the Hessian matrix
hessian_matrix = hessian(function, (x, y))
print("Hessian Matrix:")
print(hessian_matrix)
```

### 24/1/25 -- DIFFERENTIAL EQUATIONS

```python
from sympy import dsolve, Eq, Function

# Define the function and equation
y = Function('y')
x = Symbol('x')
equation = Eq(y(x).diff(x), x + y(x))

# Solve the differential equation
solution = dsolve(equation)
print("Solution of the differential equation:", solution)
```

### 25/1/25 REAL WORLD APPLICATIONS OF CALCULUS IN DATA SCIENCE

```python
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    for _ in range(iterations):
        grad = 2 * x  # Gradient of x^2
        x -= learning_rate * grad  # Update rule
    return x

# Perform gradient descent
minimum = gradient_descent(starting_point=10, learning_rate=0.1, iterations=100)
print("Minimum point:", minimum)
```
